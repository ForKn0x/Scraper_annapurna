```python
import os
import requests as req
import json
```

```python
def scraper():

    url = "https://bg.annapurnapost.com/api/search?page={}&title=सन्दीप"

    pagination = 30

    start = 0
    final_data = []

    if os.path.exists('./memory.txt'):
        with open('memory.txt', 'r') as line:
            start = int(line.readline()) + 1

    with open('news.json', 'a+', encoding='utf-8') as paper:

        for i in range(start, start + 3):
            print('PASS')

            try:
                data = req.get(url.format(i)).json()

                if req.get(url.format(i)).status_code == 200:
                    if data["status"] == 'success':

                        pagination = pagination - \
                            len(data['data'].get('items'))

                        final_data.extend(data['data']['items'])

                        if not pagination <= 0:
                            print('continue')
                        else:
                            break

            except BaseException:
                with open('memory.txt', 'w') as line:
                    line.write(str(i))
                print("error")
                break

            with open('memory.txt', 'w') as line:
                line.write(str(i))

        json.dump(final_data, paper, ensure_ascii=False, indent=4)

    paper.close()
```

```python
def checkValid():
    with open('news.json', 'r', encoding='utf-8') as paper:
        return json.load(paper)
```

```python
scraper()
```

```python
checkValid()
```
